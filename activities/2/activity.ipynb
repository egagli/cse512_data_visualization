```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
wages_df = pd.read_csv('us_wage_data.csv')
```

```python
wages_df
```

```python
df_pivot = pd.pivot_table(
	wages_df,
	values="Median Earnings (USD)",
	index="Race and ethnicity",
	columns="Occupation Group",
	aggfunc=np.mean
)
```

```python
f,ax=plt.subplots(figsize=(10,4))

df_pivot.plot(ax=ax,kind='bar',stacked=True)

f.suptitle('US Median Earnings by Race and Ethnicity')
```
